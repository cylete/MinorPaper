```python
# -*- coding: utf-8 -*-
import pandas as pd
import os

# 文件
bas_file = "FS_Combas.xlsx"
ins_file = "FS_Comins.xlsx"
scf_file = "FS_Comscfd.xlsx"
output_file = "FS_Combined.xlsx"

# 读取（只读第一个 sheet）
df_bas = pd.read_excel(bas_file, sheet_name=0)
df_ins = pd.read_excel(ins_file, sheet_name=0)
df_scf = pd.read_excel(scf_file, sheet_name=0)
```

```python
# 筛选合并报表
df_bas = df_bas[df_bas['Typrep'] == 'A'].copy()
df_ins = df_ins[df_ins['Typrep'] == 'A'].copy()
df_scf = df_scf[df_scf['Typrep'] == 'A'].copy()

print(f"筛选后行数：资产负债表 {len(df_bas)}，利润表 {len(df_ins)}，现金流量表 {len(df_scf)}")
```

```python
merge_keys = ['Stkcd', 'ShortName', 'Accper', 'Typrep', 'IfCorrect', 'DeclareDate']

# 合并
df_merged = pd.merge(df_bas, df_ins, on=merge_keys, how='outer')
df_final = pd.merge(df_merged, df_scf, on=merge_keys, how='outer')
```

```python
# 列顺序：关键字段在前
key_cols = merge_keys  # 就是上面6个
other_cols = [col for col in df_final.columns if col not in key_cols]
df_final = df_final[key_cols + other_cols]
```

```python
# 保存
df_final.to_excel(output_file, index=False, engine='openpyxl')
print(f"\n✅ 合并成功！共 {len(df_final)} 行，{len(df_final.columns)} 列")
print(f"输出文件: {os.path.abspath(output_file)}")

# 预览
financial_cols = [col for col in df_final.columns if col.startswith(('A', 'B', 'C'))][:5]
preview_cols = key_cols[:3] + financial_cols
print("\n前5行预览:")
print(df_final[preview_cols].head().to_string())
```
